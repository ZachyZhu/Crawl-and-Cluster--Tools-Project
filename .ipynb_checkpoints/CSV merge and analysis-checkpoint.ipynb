```python

df2=pd.read_csv('dataset.csv',encoding='UTF-8')
del df2['Unnamed: 0']
df2
```

```python

for i in range(len(df2['price'])):
    df2.loc[i,'price']=df2.loc[i,'price']/df2.loc[i,'area']
```

```python

index=0
high_price=0
for i in range(len(df2['price'])):
    if df2.loc[i,'price']>high_price:
        high_price=df2.loc[i,'price']
        index=i
```

```python
print(index,high_price)
```

```python

df2.loc[index,'address']
```

```python

for i in range(len(df2['price'])):
    if df2.loc[i,'price']> high_price:
        high_price=df2.loc[i,'price']
        index=i
    if df2.loc[i,'type']=='Co-op':
        df2.loc[i,'type']=1
    if df2.loc[i,'type']=='Condo':
        df2.loc[i,'type']=2
    if df2.loc[i,'type']=='Condop':
        df2.loc[i,'type']=3
    if df2.loc[i,'type']=='Multi-family':
        df2.loc[i,'type']=4
    if df2.loc[i,'type']=='Building':
        df2.loc[i,'type']=5
    if df2.loc[i,'type']=='Townhouse':
        df2.loc[i,'type']=6

patternE=r'^(\d+)\s(E)\w+\s(\d+)'
patternW=r'^(\d+)\s(W)\w+\s(\d+)'

for i in range(len(df2['price'])):
    match=re.search(patternW,str(df2.loc[i,'address']))
    if match!=None:
        df2.loc[i,'address']=int(((int(match.groups()[0])+100)*40)**2+((int(match.groups()[2])-53)*80)**2)
        
for i in range(len(df2['price'])):
    match=re.search(patternE,str(df2.loc[i,'address']))
    if match!=None:
        df2.loc[i,'address']=int(((int(match.groups()[0])-100)*40)**2+((int(match.groups()[2])-53)*80)**2)
```

```python

pd.DataFrame(df2).to_csv('processed_data.csv')
```
